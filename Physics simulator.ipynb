```python
import math
class Obj:
    def __init__(self,x,y,Vx,Vy):
        self.x = x
        self.y = y
        self.Vy = Vy
        self.Vx = Vx
class Ref:
    def __init__(self,Objs):
        self.Objs = Objs
    def update(self):
        nObjs= []
        for Ob in self.Objs:
            Xf = Ob.x + Ob.Vx
            Yf = Ob.y + Ob.Vy
            nObjs.append(Obj(Xf,Yf,Ob.Vx,Ob.Vy))
        return nObjs    
    
def Advance(Refs,T):
    for a in range(T):
        Refs.append(Ref(Refs[a].update()))
Con1 = True
Started = False
Objs = []
Refs = []
T = 0 # make it so when it goes back to T 0 all the other refrence frames after that are deleted
while Con1:
    try:
        Onum = int(input("Hello, how many Objects are there"))
        if Onum == 0:
            print("stopping sim")
            break
        print("Alright there are " + str(Onum) + " Objects")
        In = int(input("Is that correct? If not input a non-number key, else input any number"))
        for x in range(Onum):
            X = int(input("what is the starting x position for object " + str(x+1)+" "))
            Y = int(input("what is the starting y position for object " + str(x+1)+" "))
            Vx = int(input("what is the starting x Velocity for object " + str(x+1)+" "))
            Vy = int(input("what is the starting y Velocity for object " + str(x+1)+" "))
            Objs.append(Obj(X,Y,Vx,Vy))
        Refs.append(Ref(Objs))
        ObjLoaded = True
        while ObjLoaded:
            Ob = int(input("Alright which Objects values would you like to change. 0 for sim"))
            if Ob == 0:
                Time = int(input("How many seconds does it run for? "))
                Advance(Refs,Time)
                T = Time + T
                result = True
                while result:
                    Ob = (int(input("Which objects new position would you like to check")))
                    if Ob == 0:
                        result = False
                    else:
                        print(str(Refs[T].Objs[Ob-1].y)+ " is the new y position and " + str(Refs[T].Objs[Ob-1].x) + " is the new x position for Object" + str(Ob))
                    
            elif Ob > Onum:
                break
            else:
                In = int(input("Alright Which value would you like to change\n1. X position\n2. Y position\n3. X velocity\n4.Y velocity "))
                if In == 1:
                    New = int(input("Whats the new X value? "))
                    Refs[T].Objs[Ob-1].x = New
                elif In == 2:
                    New = int(input("Whats the new Y value? "))
                    Refs[T].Objs[Ob-1].y = New
                elif In == 3:
                    New = int(input("whats the new X velocity? "))
                    Refs[T].Objs[Ob-1].Vx = New
                elif In == 4:
                    New = int(input("whats the new Y velocity? "))
                    Refs[T].Objs[Ob-1].Vy = New
            
                
                
    except ValueError:
        print("Thats not a number try again")
    
    
```
